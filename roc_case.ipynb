```python
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

# Example data
y0 = [0,   1,   0,   1,   0,   1,   1,   0] 
s1 = [0.6, 0.8, 0.3, 0.9, 0.2, 0.7, 0.4, 0.4] # err 1,7
s2 = [0.3, 0.8, 0.3, 0.9, 0.2, 0.7, 0.6, 0.4] # perfect
s4 = [0.6, 0.8, 0.3, 0.9, 0.2, 0.7, 0.4, 0.6] # err 1,7,8
s3 = [0.6, 0.2, 0.3, 0.9, 0.2, 0.6, 0.2, 0.3] # err 1. 2. 7


fpr1, tpr1, _ = roc_curve(y0, s1)
auc1 = roc_auc_score(y0, s1)

fpr2, tpr2, _ = roc_curve(y0, s2)
auc2 = roc_auc_score(y0, s2)

fpr3, tpr3, _ = roc_curve(y0, s3)
auc3 = roc_auc_score(y0, s3)

fpr4, tpr4, _ = roc_curve(y0, s4)
auc4 = roc_auc_score(y0, s4)

# Print ROC-AUC Score
print(auc1,auc2,auc3,auc4)

# Plot ROC curve
plt.figure()
plt.plot(fpr1, tpr1, color='darkorange', lw=2, label=f'1')
plt.plot(fpr2, tpr2, color='navy', lw=2, linestyle='--', label=f'2')
plt.plot(fpr3, tpr3, color='red', lw=2, label=f'3')
plt.plot(fpr4, tpr4, color='grey', lw=2, label=f'4')

plt.plot([0, 1], [0, 1], color='green', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc="lower right")
plt.show()
```
